```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Align data
min_len = min(len(df_jam), len(df_djembe))
jam_prices = df_jam['mid_price'].iloc[:min_len].values.reshape(-1, 1)
squid_prices = df_squid['mid_price'].iloc[:min_len].values.reshape(-1, 1)

# Create polynomial features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(squid_prices)  # [1, x, x²]

# Fit quadratic regression
model = LinearRegression()
model.fit(X_poly, jam_prices)

# Get coefficients
intercept = model.intercept_[0]
linear_coef = model.coef_[0][1]
quadratic_coef = model.coef_[0][2]

print("Quadratic Regression Coefficients:")
print(f"Intercept (constant term): {intercept:.6f}")
print(f"Linear coefficient (SQUID): {linear_coef:.6f}")
print(f"Quadratic coefficient (SQUID²): {quadratic_coef:.6f}")

# Calculate R² score
y_pred = model.predict(X_poly)
r2 = r2_score(jam_prices, y_pred)
print(f"\nR-squared: {r2:.4f}")

# Visualize
plt.figure(figsize=(10, 6))
plt.scatter(squid_prices, jam_prices, color='blue', alpha=0.5, label='Actual Data')

# Sort for smooth curve
sort_idx = np.argsort(squid_prices.flatten())
x_plot = squid_prices[sort_idx]
y_plot = y_pred[sort_idx]

plt.plot(x_plot, y_plot, color='red', 
         label=f'Quadratic Fit: y = {quadratic_coef:.3f}x² + {linear_coef:.3f}x + {intercept:.3f}')
plt.xlabel('Squid Price')
plt.ylabel('Jam Price')
plt.title('Non-Linear Relationship: Jams vs Squid')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



# Calculate SWMID (Size-Weighted Mid Price)
def calculate_swmid(df):
    df['swmid'] = (df['bid_price_1'] * df['ask_volume_1'] + df['ask_price_1'] * df['bid_volume_1']) / \
                  (df['bid_volume_1'] + df['ask_volume_1'])
    return df

df_squid = calculate_swmid(df_squid)
df_jam = calculate_swmid(df_jam)

# Align timestamps
combined = pd.DataFrame({
    'timestamp': df_jam['timestamp'],
    'jams_swmid': df_jam['swmid'],
    'squid_swmid': df_squid['swmid']
})

# Spread = Jams + Squid (for negative correlation)
combined['spread'] = combined['jams_swmid'] + combined['squid_swmid']

# Plot spread
plt.figure(figsize=(15, 5))
plt.plot(combined['timestamp'], combined['spread'], label='Spread (Jams + Squid)', color='purple')
plt.axhline(combined['spread'].mean(), color='r', linestyle='--', label='Mean')
plt.axhline(combined['spread'].mean() + 2*combined['spread'].std(), color='g', linestyle=':', label='±2σ')
plt.axhline(combined['spread'].mean() - 2*combined['spread'].std(), color='g', linestyle=':')
plt.title('Jams-Squid Spread (Negative Correlation)')
plt.xlabel('Timestamp')
plt.ylabel('Spread Value')
plt.legend()
plt.grid()
plt.show()

# Trading signals
combined['z_score'] = (combined['spread'] - combined['spread'].mean()) / combined['spread'].std()
combined['signal'] = np.where(
    combined['z_score'] > 2, -1,  # Sell Jams + Buy Squid
    np.where(combined['z_score'] < -2, 1, 0)  # Buy Jams + Sell Squid
)

# Print stats
print(f"Spread Statistics:")
print(f"Mean: {combined['spread'].mean():.2f}")
print(f"Std Dev: {combined['spread'].std():.2f}")
print(f"Correlation: -0.80 (strongly negative)")

# Show signals
print("\nSample Signals:")
print(combined[['timestamp', 'jams_swmid', 'squid_swmid', 'spread', 'z_score', 'signal']].head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Compute Mid-Price
df_crois['mid_price'] = (df_crois['bid_price_1'] + df_crois['ask_price_1']) / 2

# Bollinger Bands Calculation
window = 20  # Moving average window
df_crois['SMA'] = df_crois['mid_price'].rolling(window=window).mean()
df_crois['STD'] = df_crois['mid_price'].rolling(window=window).std()
df_crois['Upper_BB'] = df_crois['SMA'] + 2 * df_crois['STD']
df_crois['Lower_BB'] = df_crois['SMA'] - 2 * df_crois['STD']

# Plotting
plt.figure(figsize=(12,6))
plt.plot(df_crois['mid_price'], label='Mid-Price', color='blue')
plt.plot(df_crois['SMA'], label='SMA', color='black')
plt.plot(df_crois['Upper_BB'], label='Upper Bollinger Band', linestyle='dashed', color='red')
plt.plot(df_crois['Lower_BB'], label='Lower Bollinger Band', linestyle='dashed', color='green')

plt.title("Mid-Price with Bollinger Bands")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

# RSI Calculation
delta = df_crois['mid_price'].diff()
gain = np.where(delta > 0, delta, 0)
loss = np.where(delta < 0, -delta, 0)

roll_up = pd.Series(gain).rolling(window=14).mean()
roll_down = pd.Series(loss).rolling(window=14).mean()

RS = roll_up / roll_down
df_crois['RSI'] = 100 - (100 / (1 + RS))

# Plot RSI
plt.figure(figsize=(12,4))
plt.plot(df_crois['RSI'], label='RSI', color='purple')
plt.axhline(70, color='red', linestyle='dashed', label='Overbought (70)')
plt.axhline(30, color='green', linestyle='dashed', label='Oversold (30)')
plt.title("Relative Strength Index (RSI)")
plt.xlabel("Time")
plt.ylabel("RSI")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

# Parameters
threshold = 0.2  # Adjust this based on your market's volatility

# Generate signals
df_crois['Buy_Signal'] = np.where((df_crois['MACD_Histogram'] > threshold) & 
                                 (df_crois['MACD_Histogram'].shift(1) <= threshold), 
                                 df_crois['mid_price'], np.nan)

df_crois['Sell_Signal'] = np.where((df_crois['MACD_Histogram'] < -threshold) & 
                                  (df_crois['MACD_Histogram'].shift(1) >= -threshold), 
                                  df_crois['mid_price'], np.nan)

# Create figure with subplots
plt.figure(figsize=(14, 10))

# First subplot: Price with buy/sell signals
plt.subplot(2, 1, 1)
plt.plot(df_crois['mid_price'], label='Mid Price', color='black', alpha=0.8, linewidth=1.5)
plt.scatter(df_crois.index, df_crois['Buy_Signal'], label='Buy Signal', 
            marker='^', color='green', s=100)
plt.scatter(df_crois.index, df_crois['Sell_Signal'], label='Sell Signal', 
            marker='v', color='red', s=100)
plt.title(f"Mid Price with MACD Histogram Signals (Threshold = ±{threshold})")
plt.ylabel("Price")
plt.legend()
plt.grid()

# Second subplot: MACD components
plt.subplot(2, 1, 2)
plt.plot(df_crois['MACD'], label='MACD', color='blue', alpha=0.7)
plt.plot(df_crois['Signal_Line'], label='Signal Line', color='orange', alpha=0.7)
plt.bar(df_crois.index, df_crois['MACD_Histogram'], 
        label='MACD Histogram', color='gray', alpha=0.5)

# Add threshold lines
plt.axhline(threshold, color='green', linestyle='--', alpha=0.5, label='Buy Threshold')
plt.axhline(-threshold, color='red', linestyle='--', alpha=0.5, label='Sell Threshold')

plt.title("MACD Components")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
