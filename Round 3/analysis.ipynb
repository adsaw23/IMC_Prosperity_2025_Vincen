```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np




def plot_acceleration_signals(df_crois):
    # Calculate velocity and acceleration
    df_crois['velocity'] = df_crois['mid_price'].diff()
    df_crois['acceleration'] = df_crois['velocity'].diff()
    
    # Generate signals
    df_crois['signal'] = np.where(
        df_crois['acceleration'] > 0.1, 
        'buy',
        np.where(
            df_crois['acceleration'] < -0.1,
            'sell',
            'hold'
        )
    )
    
    # Create plot
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 10), sharex=True)
    
    # Price plot
    ax1.plot(df_crois['timestamp'], df_crois['mid_price'], label='Price', color='royalblue')
    ax1.scatter(
        df_crois[df_crois['signal'] == 'buy']['timestamp'],
        df_crois[df_crois['signal'] == 'buy']['mid_price'],
        color='green', marker='^', s=100, label='Buy'
    )
    ax1.scatter(
        df_crois[df_crois['signal'] == 'sell']['timestamp'],
        df_crois[df_crois['signal'] == 'sell']['mid_price'],
        color='red', marker='v', s=100, label='Sell'
    )
    ax1.set_ylabel('Price')
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)
    
    # Velocity plot
    ax2.plot(df_crois['timestamp'], df_crois['velocity'], label='Velocity', color='purple')
    ax2.axhline(0, color='gray', linestyle='--')
    ax2.set_ylabel('Velocity (1st deriv)')
    ax2.grid(True, linestyle='--', alpha=0.7)
    
    # Acceleration plot
    ax3.plot(df_crois['timestamp'], df_crois['acceleration'], label='Acceleration', color='orange')
    ax3.axhline(0.1, color='green', linestyle=':', label='Buy Threshold')
    ax3.axhline(-0.1, color='red', linestyle=':', label='Sell Threshold')
    ax3.axhline(0, color='gray', linestyle='--')
    ax3.set_ylabel('Acceleration (2nd deriv)')
    ax3.set_xlabel('Timestamp')
    ax3.legend()
    ax3.grid(True, linestyle='--', alpha=0.7)
    
    plt.suptitle('CROISSANT Price Acceleration Strategy', y=0.92)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Example usage:
plot_acceleration_signals(df_crois)
```

```python
def plot_velocity_colored_signals(df_crois):
    # Calculate velocity
    df_crois['velocity'] = df_crois['mid_price'].diff()
    
    # Create segments for colored line
    df_crois['segment_color'] = np.where(
        df_crois['velocity'] > 0.2,
        'green',
        np.where(
            df_crois['velocity'] < -0.2,
            'red',
            'gray'
        )
    )
    
    # Create the plot
    plt.figure(figsize=(15, 7))
    
    # Plot colored price line segment by segment
    for i in range(1, len(df_crois)):
        plt.plot(
            df_crois['timestamp'].iloc[i-1:i+1],
            df_crois['mid_price'].iloc[i-1:i+1],
            color=df_crois['segment_color'].iloc[i],
            linewidth=2
        )
    
    # Add threshold reference lines
    plt.axhline(y=df_crois['mid_price'].mean(), color='blue', linestyle='--', alpha=0.5, label='Mean Price')
    
    # Create custom legend
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], color='green', lw=2, label='Velocity > +1 (Strong Up)'),
        Line2D([0], [0], color='red', lw=2, label='Velocity < -1 (Strong Down)'),
        Line2D([0], [0], color='gray', lw=2, label='Steady (-1 ≤ Velocity ≤ +1)'),
        Line2D([0], [0], color='blue', linestyle='--', lw=1, label='Mean Price')
    ]
    
    # Formatting
    plt.title('CROISSANT Price with Velocity Coloring', fontsize=16, pad=20)
    plt.xlabel('Timestamp', fontsize=12)
    plt.ylabel('Price', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(handles=legend_elements, loc='upper left')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Example usage:
plot_velocity_colored_signals(df_crois)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load all datasets
products = {
    'kelp': pd.read_csv("../data/round_4/kelp_data.csv"),
    'resin': pd.read_csv("../data/round_4/resin_data.csv"),
    'squid': pd.read_csv("../data/round_4/squid_data.csv"),
    'jam': pd.read_csv("../data/round_4/jams_data.csv"),
    'djembe': pd.read_csv("../data/round_4/djembe_data.csv"),
    'croissant': pd.read_csv("../data/round_4/croissant_data.csv"),
    'basket_1': pd.read_csv("../data/round_4/basket_1_data.csv"),
    'basket_2': pd.read_csv("../data/round_4/basket_2_data.csv"),
    'rock': pd.read_csv("../data/round_4/rock_data.csv"),
    'voucher_9500': pd.read_csv("../data/round_4/voucher_9500_data.csv"),
    'voucher_9750': pd.read_csv("../data/round_4/voucher_9750_data.csv"),
    'voucher_10000': pd.read_csv("../data/round_4/voucher_10000_data.csv"),
    'voucher_10250': pd.read_csv("../data/round_4/voucher_10250_data.csv"),
    'voucher_10500': pd.read_csv("../data/round_4/voucher_10500_data.csv")
}

# Function to extract mid prices and align by timestamp
def prepare_correlation_data(products):
    # Create a base dataframe with all timestamps
    all_timestamps = set()
    for df in products.values():
        if 'timestamp' in df.columns:
            all_timestamps.update(df['timestamp'].unique())
    
    # Create a consolidated dataframe
    correlation_df = pd.DataFrame(index=sorted(all_timestamps))
    
    # Add mid prices for each product
    for name, df in products.items():
        if 'timestamp' not in df.columns:
            continue
            
        # Calculate mid price (assuming we have bid/ask columns)
        if 'bid_price_1' in df.columns and 'ask_price_1' in df.columns:
            df['mid_price'] = (df['bid_price_1'] + df['ask_price_1']) / 2
        elif 'price' in df.columns:
            df['mid_price'] = df['price']
        else:
            continue
            
        # Add to correlation df
        temp_df = df[['timestamp', 'mid_price']].drop_duplicates('timestamp')
        temp_df = temp_df.set_index('timestamp')
        correlation_df[name] = temp_df['mid_price']
    
    return correlation_df.dropna(how='all')

# Prepare correlation data
correlation_df = prepare_correlation_data(products)

# Calculate correlations
correlation_matrix = correlation_df.corr()

# Visualize the correlation matrix
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title("Correlation Matrix of All Products")
plt.tight_layout()
plt.show()

# Print strongest correlations (absolute value > 0.7)
print("\nStrong Correlations (|r| > 0.7):")
corr_pairs = correlation_matrix.unstack()
strong_corrs = corr_pairs[abs(corr_pairs) > 0.7].sort_values(ascending=False)
strong_corrs = strong_corrs[strong_corrs < 1.0]  # Remove self-correlations
print(strong_corrs)
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import brentq
from scipy.stats import norm
import matplotlib.pyplot as plt

# You already have this part to load your data
# df_rock = pd.read_csv("../data/round_3/rock_data.csv")
# df_v9500 = pd.read_csv("../data/round_3/voucher_9500_data.csv")
# df_v9750 = pd.read_csv("../data/round_3/voucher_9750_data.csv")
# df_v10000 = pd.read_csv("../data/round_3/voucher_10000_data.csv")
# df_v10250 = pd.read_csv("../data/round_3/voucher_10250_data.csv")
# df_v10500 = pd.read_csv("../data/round_3/voucher_10500_data.csv")

# You already have this function to calculate mid prices
# def get_mid_price(df):
#     df['mid_price'] = (df['bid_price_1'] + df['ask_price_1'])/2
#     return df
# dfs = {
#     'ROCK': get_mid_price(df_rock),
#     'V9500': get_mid_price(df_v9500),
#     'V9750': get_mid_price(df_v9750),
#     'V10000': get_mid_price(df_v10000),
#     'V10250': get_mid_price(df_v10250),
#     'V10500': get_mid_price(df_v10500)
# }

# Define Black-Scholes formula for call options
def black_scholes_call(S, K, T, r, sigma):
    """
    Calculate Black-Scholes price for a call option
    
    Parameters:
    S: Spot price of the underlying asset
    K: Strike price
    T: Time to expiration (in years)
    r: Risk-free interest rate (decimal)
    sigma: Volatility (decimal)
    
    Returns:
    call_price: Price of the call option
    """
    if T <= 0:
        # If option expired, check if it has value
        return max(0, S - K)
    
    d1 = (np.log(S/K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

# Function to calculate implied volatility using root finding
def implied_volatility(price, S, K, T, r):
    """
    Calculate implied volatility using the Brent method
    
    Parameters:
    price: Market price of the option
    S: Spot price of the underlying asset
    K: Strike price
    T: Time to expiration (in years)
    r: Risk-free interest rate (decimal)
    
    Returns:
    sigma: Implied volatility
    """
    # Define the objective function (difference between market price and BS price)
    def objective(sigma):
        return black_scholes_call(S, K, T, r, sigma) - price
    
    # Check if option is significantly out of the money
    if price < 0.01:
        return np.nan
    
    try:
        # Use Brent method to find implied volatility (search in range 0.001 to 5.0)
        # Typical volatility is between 0.1 (10%) and 1.0 (100%)
        implied_vol = brentq(objective, 0.001, 5.0, maxiter=1000)
        return implied_vol
    except:
        # If root finding fails, return NaN
        return np.nan

# Function to analyze data and calculate implied volatility for each voucher
def calculate_implied_volatilities(dfs, current_round=3):
    """
    Calculate implied volatilities for each voucher
    
    Parameters:
    dfs: Dictionary of dataframes with price data
    current_round: Current round number (1-7)
    
    Returns:
    results_df: DataFrame with implied volatilities
    """
    # Risk-free interest rate (assumed, can be adjusted)
    r = 0.02  # 2%
    
    # Time to expiration in years (assuming 7 trading days at the start)
    # At round 3, there would be 7 - (3-1) = 5 days left
    days_to_expiry = 7 - (current_round - 1)
    T = days_to_expiry / 365  # Convert to years
    
    # Initialize a list to store results
    results = []
    
    # Get the most recent spot price of the underlying asset
    spot_price = dfs['ROCK']['mid_price'].iloc[-1]
    
    # Mapping of dataframe keys to strike prices
    strike_prices = {
        'V9500': 9500,
        'V9750': 9750,
        'V10000': 10000,
        'V10250': 10250,
        'V10500': 10500
    }
    
    # Calculate implied volatility for each voucher
    for voucher_key, strike_price in strike_prices.items():
        # Get the most recent mid price of the voucher
        option_price = dfs[voucher_key]['mid_price'].iloc[-1]
        
        # Calculate implied volatility
        iv = implied_volatility(option_price, spot_price, strike_price, T, r)
        
        # Store the results
        results.append({
            'Voucher': voucher_key,
            'Strike': strike_price,
            'Option_Price': option_price,
            'Spot_Price': spot_price,
            'Days_To_Expiry': days_to_expiry,
            'Implied_Volatility': iv,
            'Implied_Volatility_Percent': iv * 100 if not np.isnan(iv) else np.nan
        })
    
    # Convert results to DataFrame
    results_df = pd.DataFrame(results)
    
    # Calculate mean implied volatility (excluding NaN values)
    mean_iv = results_df['Implied_Volatility'].mean(skipna=True)
    print(f"Mean Implied Volatility: {mean_iv:.4f} ({mean_iv*100:.2f}%)")
    
    return results_df

# Calculate implied volatilities (change the current_round parameter as needed)
iv_results = calculate_implied_volatilities(dfs, current_round=3)

# Display the results
display(iv_results)

# Plot the implied volatility smile
plt.figure(figsize=(10, 6))
plt.plot(iv_results['Strike'], iv_results['Implied_Volatility_Percent'], 'o-', linewidth=2)
plt.grid(True, alpha=0.3)
plt.title('Implied Volatility Smile for Volcanic Rock Vouchers')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility (%)')
plt.axhline(y=iv_results['Implied_Volatility_Percent'].mean(), color='r', linestyle='--', label=f'Mean IV: {iv_results["Implied_Volatility_Percent"].mean():.2f}%')
plt.legend()
plt.tight_layout()
plt.show()

# Function to calculate implied volatility over time for each voucher
def calculate_implied_volatility_series(dfs, voucher_key, strike_price):
    """
    Calculate implied volatility series for a specific voucher
    
    Parameters:
    dfs: Dictionary of dataframes with price data
    voucher_key: Key of the voucher dataframe
    strike_price: Strike price of the voucher
    
    Returns:
    iv_series: Series of implied volatilities over time
    """
    # Risk-free interest rate (assumed, can be adjusted)
    r = 0.02  # 2%
    
    # Get spot prices and option prices
    spot_prices = dfs['ROCK']['mid_price']
    option_prices = dfs[voucher_key]['mid_price']
    
    # Ensure data is aligned
    min_length = min(len(spot_prices), len(option_prices))
    spot_prices = spot_prices[:min_length]
    option_prices = option_prices[:min_length]
    
    # Calculate time index (assuming data is in chronological order)
    time_indices = list(range(min_length))
    
    # Initialize list to store implied volatilities
    iv_values = []
    
    # Calculate implied volatility for each time point
    for i in time_indices:
        # Assuming we start in round 3 with 5 days to expiry
        # and each time point represents a fraction of a day
        # This is a simplified approach - you may need to adjust for your actual data
        days_to_expiry = 5 - (i / (min_length-1))  # Linear decrease from 5 to 0
        T = max(days_to_expiry, 0.01) / 365  # Minimum of 0.01 days to avoid division by zero
        
        iv = implied_volatility(option_prices.iloc[i], spot_prices.iloc[i], strike_price, T, r)
        iv_values.append(iv)
    
    return pd.Series(iv_values, index=time_indices)

# Calculate and plot implied volatility over time for each voucher
plt.figure(figsize=(12, 8))

for voucher_key, strike_price in {'V9500': 9500, 'V9750': 9750, 'V10000': 10000, 'V10250': 10250, 'V10500': 10500}.items():
    iv_series = calculate_implied_volatility_series(dfs, voucher_key, strike_price)
    plt.plot(iv_series.index, iv_series * 100, label=f'{voucher_key} (Strike={strike_price})')

plt.grid(True, alpha=0.3)
plt.title('Implied Volatility Over Time for Different Volcanic Rock Vouchers')
plt.xlabel('Time Index')
plt.ylabel('Implied Volatility (%)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
class BlackScholes:
    @staticmethod
    def black_scholes_call(spot, strike, time_to_expiry, volatility):
        d1 = (
            log(spot) - log(strike) + (0.5 * volatility * volatility) * time_to_expiry
        ) / (volatility * sqrt(time_to_expiry))
        d2 = d1 - volatility * sqrt(time_to_expiry)
        call_price = spot * NormalDist().cdf(d1) - strike * NormalDist().cdf(d2)
        return call_price

    @staticmethod
    def black_scholes_put(spot, strike, time_to_expiry, volatility):
        d1 = (log(spot / strike) + (0.5 * volatility * volatility) * time_to_expiry) / (
            volatility * sqrt(time_to_expiry)
        )
        d2 = d1 - volatility * sqrt(time_to_expiry)
        put_price = strike * NormalDist().cdf(-d2) - spot * NormalDist().cdf(-d1)
        return put_price

    @staticmethod
    def delta(spot, strike, time_to_expiry, volatility):
        d1 = (
            log(spot) - log(strike) + (0.5 * volatility * volatility) * time_to_expiry
        ) / (volatility * sqrt(time_to_expiry))
        return NormalDist().cdf(d1)

    @staticmethod
    def gamma(spot, strike, time_to_expiry, volatility):
        d1 = (
            log(spot) - log(strike) + (0.5 * volatility * volatility) * time_to_expiry
        ) / (volatility * sqrt(time_to_expiry))
        return NormalDist().pdf(d1) / (spot * volatility * sqrt(time_to_expiry))

    @staticmethod
    def vega(spot, strike, time_to_expiry, volatility):
        d1 = (
            log(spot) - log(strike) + (0.5 * volatility * volatility) * time_to_expiry
        ) / (volatility * sqrt(time_to_expiry))
        # print(f"d1: {d1}")
        # print(f"vol: {volatility}")
        # print(f"spot: {spot}")
        # print(f"strike: {strike}")
        # print(f"time: {time_to_expiry}")
        return NormalDist().pdf(d1) * (spot * sqrt(time_to_expiry)) / 100

    @staticmethod
    def implied_volatility(
        call_price, spot, strike, time_to_expiry, max_iterations=200, tolerance=1e-10
    ):
        low_vol = 0.01
        high_vol = 1.0
        volatility = (low_vol + high_vol) / 2.0  # Initial guess as the midpoint
        for _ in range(max_iterations):
            estimated_price = BlackScholes.black_scholes_call(
                spot, strike, time_to_expiry, volatility
            )
            diff = estimated_price - call_price
            if abs(diff) < tolerance:
                break
            elif diff > 0:
                high_vol = volatility
            else:
                low_vol = volatility
            volatility = (low_vol + high_vol) / 2.0
        return volatility


```

```python


def calculate_iv(row, strike, underlying_df):
    try:
        S = underlying_df.loc[row.name, 'mid_price']  # Underlying price
        V = row['mid_price']  # Voucher price
        return BlackScholes.implied_volatility(V, S, strike, tte)
    except:
        return np.nan

# Add IV to each voucher dataframe
for strike, df_name in [(9500, 'V9500'), 
                       (9750, 'V9750'),
                       (10000, 'V10000'),
                       (10250, 'V10250'),
                       (10500, 'V10500')]:
    
    # Ensure we have matching timestamps
    aligned_rock = dfs['ROCK'].set_index('timestamp').reindex(dfs[df_name]['timestamp'])
    dfs[df_name]['implied_vol'] = dfs[df_name].apply(
        lambda row: calculate_iv(row, strike, aligned_rock), 
        axis=1
    )

# Calculate mean volatility for each strike
mean_volatilities = {}
for strike, df_name in [(9500, 'V9500'), 
                       (9750, 'V9750'),
                       (10000, 'V10000'),
                       (10250, 'V10250'),
                       (10500, 'V10500')]:
    
    # Remove outliers (IV typically between 5% and 200%)
    clean_iv = dfs[df_name]['implied_vol'][
        (dfs[df_name]['implied_vol'] > 0.05) & 
        (dfs[df_name]['implied_vol'] < 2.00)
    ]
    
    mean_vol = clean_iv.mean()
    mean_volatilities[strike] = mean_vol
    print(f"Mean IV for {strike}: {mean_vol:.4f}")

# Overall mean volatility
all_iv = pd.concat([dfs[df_name]['implied_vol'] for df_name in ['V9500', 'V9750', 'V10000', 'V10250', 'V10500']])
clean_all_iv = all_iv[(all_iv > 0.05) & (all_iv < 2.00)]
print(f"\nOverall mean IV: {clean_all_iv.mean():.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



# Calculate SWMID (Size-Weighted Mid Price)
def calculate_swmid(df):
    df['swmid'] = (df['bid_price_1'] * df['ask_volume_1'] + df['ask_price_1'] * df['bid_volume_1']) / \
                  (df['bid_volume_1'] + df['ask_volume_1'])
    return df

df_squid = calculate_swmid(df_squid)
df_jam = calculate_swmid(df_jam)

# Align timestamps
combined = pd.DataFrame({
    'timestamp': df_jam['timestamp'],
    'jams_swmid': df_jam['swmid'],
    'squid_swmid': df_squid['swmid']
})

# Spread = Jams + Squid (for negative correlation)
combined['spread'] = combined['jams_swmid'] + combined['squid_swmid']

# Plot spread
plt.figure(figsize=(15, 5))
plt.plot(combined['timestamp'], combined['spread'], label='Spread (Jams + Squid)', color='purple')
plt.axhline(combined['spread'].mean(), color='r', linestyle='--', label='Mean')
plt.axhline(combined['spread'].mean() + 2*combined['spread'].std(), color='g', linestyle=':', label='±2σ')
plt.axhline(combined['spread'].mean() - 2*combined['spread'].std(), color='g', linestyle=':')
plt.title('Jams-Squid Spread (Negative Correlation)')
plt.xlabel('Timestamp')
plt.ylabel('Spread Value')
plt.legend()
plt.grid()
plt.show()

# Trading signals
combined['z_score'] = (combined['spread'] - combined['spread'].mean()) / combined['spread'].std()
combined['signal'] = np.where(
    combined['z_score'] > 2, -1,  # Sell Jams + Buy Squid
    np.where(combined['z_score'] < -2, 1, 0)  # Buy Jams + Sell Squid
)

# Print stats
print(f"Spread Statistics:")
print(f"Mean: {combined['spread'].mean():.2f}")
print(f"Std Dev: {combined['spread'].std():.2f}")
print(f"Correlation: -0.80 (strongly negative)")

# Show signals
print("\nSample Signals:")
print(combined[['timestamp', 'jams_swmid', 'squid_swmid', 'spread', 'z_score', 'signal']].head())
```
