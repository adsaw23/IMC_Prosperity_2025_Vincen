```python
import pandas as pd
import matplotlib.pyplot as plt

# Compute Mid-Price
df_squid['mid_price'] = (df_squid['bid_price_1'] + df_squid['ask_price_1']) / 2

# Bollinger Bands Calculation
window = 20  # Moving average window
df_squid['SMA'] = df_squid['mid_price'].rolling(window=window).mean()
df_squid['STD'] = df_squid['mid_price'].rolling(window=window).std()
df_squid['Upper_BB'] = df_squid['SMA'] + 2 * df_squid['STD']
df_squid['Lower_BB'] = df_squid['SMA'] - 2 * df_squid['STD']

# Plotting
plt.figure(figsize=(12,6))
plt.plot(df_squid['mid_price'], label='Mid-Price', color='blue')
plt.plot(df_squid['SMA'], label='SMA', color='black')
plt.plot(df_squid['Upper_BB'], label='Upper Bollinger Band', linestyle='dashed', color='red')
plt.plot(df_squid['Lower_BB'], label='Lower Bollinger Band', linestyle='dashed', color='green')

plt.title("Mid-Price with Bollinger Bands")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()
```

```python
df_squid['spread'] = np.where(
    (df_squid['bid_price_1'] == 0) | (df_squid['ask_price_1'] == 0),
    0,
    df_squid['ask_price_1'] - df_squid['bid_price_1']
)
mean_spread = df_squid[df_squid['spread'] != 0]['spread'].mean()
mean_spread
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Compute Mid-Price and Spread
df_squid['mid_price'] = (df_squid['bid_price_1'] + df_squid['ask_price_1']) / 2
df_squid['spread'] = np.where(
    (df_squid['bid_price_1'] == 0) | (df_squid['ask_price_1'] == 0),
    0,
    df_squid['ask_price_1'] - df_squid['bid_price_1']
)

# Indices where spread is high or low
spread_high = df_squid['spread'] >= 4
spread_low = df_squid['spread'] <= 1

# Plotting side-by-side
fig, axs = plt.subplots(1, 2, figsize=(14, 5), sharex=True)

# Plot Mid-Price
axs[0].plot(df_squid['mid_price'], label='Mid-Price', color='blue')
axs[0].scatter(df_squid.index[spread_high], df_squid['mid_price'][spread_high], color='red', label='Spread >= 4', marker='o')
axs[0].scatter(df_squid.index[spread_low], df_squid['mid_price'][spread_low], color='green', label='Spread <= 1', marker='x')
axs[0].set_title('Mid-Price Over Time')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Price')
axs[0].legend()
axs[0].grid()

# Plot Spread
axs[1].plot(df_squid['spread'], label='Spread', color='purple')
axs[1].set_title('Bid-Ask Spread Over Time')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Spread')
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming df_squid has columns like:
# ['bid_volume_1', 'bid_volume_2', 'bid_volume_3', 
#  'ask_volume_1', 'ask_volume_2', 'ask_volume_3']

# Create figure with 6 subplots (3 bid, 3 ask)
plt.figure(figsize=(15, 10))

# Plot bid volumes
for i in range(1, 4):
    plt.subplot(2, 3, i)
    plt.hist(df_squid[f'bid_volume_{i}'], 
             bins=30, 
             color='blue', 
             alpha=0.7)
    plt.title(f'Bid Volume {i} Distribution')
    plt.xlabel('Volume')
    plt.ylabel('Frequency')
    plt.grid(True)

# Plot ask volumes (absolute values since ask volumes are negative)
for i in range(1, 4):
    plt.subplot(2, 3, i+3)
    plt.hist(df_squid[f'ask_volume_{i}'].abs(), 
             bins=30, 
             color='red', 
             alpha=0.7)
    plt.title(f'Ask Volume {i} Distribution')
    plt.xlabel('Volume (Absolute)')
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()

# Print volume statistics
print("Volume Statistics:")
for vol_type in ['bid', 'ask']:
    for i in range(1, 4):
        col = f'{vol_type}_volume_{i}'
        abs_vals = df_squid[col].abs() if vol_type == 'ask' else df_squid[col]
        print(f"\n{col.upper()}:")
        print(f"Mean: {abs_vals.mean():.2f}")
        print(f"Median: {abs_vals.median():.2f}")
        print(f"Std Dev: {abs_vals.std():.2f}")
        print(f"Max: {abs_vals.max()}")
        print(f"Min: {abs_vals.min()}")
```
